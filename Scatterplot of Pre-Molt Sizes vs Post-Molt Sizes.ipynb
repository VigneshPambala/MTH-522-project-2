```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the stored file (mp2 - input.csv)
df = pd.read_csv('input.csv')

# Create a scatterplot of pre-molt sizes versus post-molt sizes
plt.figure(figsize=(8, 6))
plt.scatter(df[df['Location'] == 'Lab']['Post-molt'], df[df['Location'] == 'Lab']['Pre-molt'], color='blue', alpha=0.6)

# Adding labels and title
plt.title('Scatterplot of Pre-Molt Sizes vs Post-Molt Sizes for Lab data', fontsize=14)
plt.xlabel('Post-Molt Size', fontsize=12)
plt.ylabel('Pre-Molt Size', fontsize=12)

# Show the plot
plt.grid(True)
plt.show()

# Create a scatterplot of pre-molt sizes versus post-molt sizes
plt.figure(figsize=(8, 6))
plt.scatter(df[df['Location'] == 'Field']['Post-molt'], df[df['Location'] == 'Field']['Pre-molt'], color='blue', alpha=0.6)

# Adding labels and title
plt.title('Scatterplot of Pre-Molt Sizes vs Post-Molt Sizes for Field data', fontsize=14)
plt.xlabel('Post-Molt Size', fontsize=12)
plt.ylabel('Pre-Molt Size', fontsize=12)

# Show the plot
plt.grid(True)
plt.show()
```

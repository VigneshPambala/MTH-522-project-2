```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'input.csv'
data = pd.read_csv(file_path)


postmolt = data[data['Location'] == 'Lab']['Post-molt']
premolt = data[data['Location'] == 'Lab']['Pre-molt']


df = pd.DataFrame({
    'premolt': premolt,
    'postmolt': postmolt
})


plt.figure(figsize=(8, 6))
sns.scatterplot(x='premolt', y='postmolt', data=df)


plt.title('Pre-molt vs Post-molt Size for Lab Crabs for Lab')
plt.xlabel('Pre-molt Size')
plt.ylabel('Post-molt Size')
plt.show()

data = pd.read_csv(file_path)


postmolt = data[data['Location'] == 'Field']['Post-molt']
premolt = data[data['Location'] == 'Field']['Pre-molt']


df = pd.DataFrame({
    'premolt': premolt,
    'postmolt': postmolt
})


plt.figure(figsize=(8, 6))
sns.scatterplot(x='premolt', y='postmolt', data=df)


plt.title('Pre-molt vs Post-molt Size for Lab Crabs for Field')
plt.xlabel('Pre-molt Size')
plt.ylabel('Post-molt Size')
plt.show()
```

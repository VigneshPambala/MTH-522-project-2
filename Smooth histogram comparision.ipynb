```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the stored file (mp2 - input.csv)
df = pd.read_csv('input.csv')

# Plot smooth histograms (KDE) for pre-molt and post-molt sizes
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['Location'] == 'Lab']['Pre-molt'], label='Pre-Molt Size', shade=True, color='blue')
sns.kdeplot(df[df['Location'] == 'Lab']['Post-molt'], label='Post-Molt Size', shade=True, color='green')

# Adding labels and title
plt.title('Comparison of Pre-Molt and Post-Molt Size Distributions for Lab Data', fontsize=14)
plt.xlabel('Size', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['Location'] == 'Field']['Pre-molt'], label='Pre-Molt Size', shade=True, color='blue')
sns.kdeplot(df[df['Location'] == 'Field']['Post-molt'], label='Post-Molt Size', shade=True, color='green')

# Adding labels and title
plt.title('Comparison of Pre-Molt and Post-Molt Size Distributions for Field Data', fontsize=14)
plt.xlabel('Size', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.show()
```

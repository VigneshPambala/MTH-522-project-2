```python
import pandas as pd
df = pd.read_csv('input.csv')



max_value_premolt_lab =df[df['Location'] == 'Lab']['Pre-molt'].max()
min_value_premolt_lab =df[df['Location'] == 'Lab']['Pre-molt'].min()
max_value_postmolt_lab =df[df['Location'] == 'Lab']['Post-molt'].max()
min_value_postmolt_lab =df[df['Location'] == 'Lab']['Post-molt'].min()


max_value_premolt_field =df[df['Location'] == 'Field']['Pre-molt'].max()
min_value_premolt_field =df[df['Location'] == 'Field']['Pre-molt'].min()
max_value_postmolt_field =df[df['Location'] == 'Field']['Post-molt'].max()
min_value_postmolt_field =df[df['Location'] == 'Field']['Post-molt'].min()

print(f'Maximum values of Pre-molt and Post-molt for Lab data:\n{max_value_premolt_lab} , {max_value_postmolt_lab}')
print(f'Minimum values of Pre-molt and Post-molt for Lab data:\n{min_value_premolt_lab} , {min_value_postmolt_lab}')

print(f'\n')


print(f'Maximum values of Pre-molt and Post-molt for Field data:\n{max_value_premolt_field} , {max_value_postmolt_field}')
print(f'Minimum values of Pre-molt and Post-molt for Field data :\n{min_value_premolt_field} , {min_value_postmolt_field}')


print(f'----------------------------------------------------------------------------------------------------------------------------------------- ')

mean_Pre = df[df['Location'] == 'Lab']['Pre-molt'].mean()
mean_Post = df[df['Location'] == 'Lab']['Post-molt'].mean()
print(f'Pre-molt mean for Lab: {mean_Pre} \nPost-molt mean Lab: {mean_Post} ')

mean_Pre = df[df['Location'] == 'Field']['Pre-molt'].mean()
mean_Post = df[df['Location'] == 'Field']['Post-molt'].mean()
print(f'\nPre-molt mean Field: {mean_Pre} \nPost-molt mean Field: {mean_Post} ')

print(f'----------------------------------------------------------------------------------------------------------------------------------------- ')

mean_Pre = df[df['Location'] == 'Lab']['Pre-molt'].median()
mean_Post = df[df['Location'] == 'Lab']['Post-molt'].median()
print(f'Pre-molt Medain for Lab: {mean_Pre} \nPost-molt Median Lab: {mean_Post} ')

mean_Pre = df[df['Location'] == 'Field']['Pre-molt'].median()
mean_Post = df[df['Location'] == 'Field']['Post-molt'].median()
print(f'\nPre-molt Median Field: {mean_Pre} \nPost-molt Median Field: {mean_Post} ')


print(f'----------------------------------------------------------------------------------------------------------------------------------------- ')


std_Pre_lab= df[df['Location'] == 'Lab']['Pre-molt'].std()
std_Post_lab =df[df['Location'] == 'Lab']['Post-molt'].std()

std_Pre_field= df[df['Location'] == 'Field']['Pre-molt'].std()
std_Post_field =df[df['Location'] == 'Field']['Post-molt'].std()
print(f'Pre-molt Standard Deviation: {std_Pre_lab} \nPost-molt Standard Deviation: {std_Post_lab} ')
print(f'\n')
print(f'Pre-molt Standard Deviation: {std_Pre_field} \nPost-molt Standard Deviation: {std_Post_field} ')

print(f'----------------------------------------------------------------------------------------------------------------------------------------- ')


skew_Pre_lab = df[df['Location'] == 'Lab']['Pre-molt'].skew()
skew_Post_field =df[df['Location'] == 'Field']['Post-molt'].skew()
skew_Post_lab =df[df['Location'] == 'Lab']['Post-molt'].skew()
skew_Pre_field = df[df['Location'] == 'Field']['Pre-molt'].skew()
print(f'Pre-molt skewness: {skew_Pre_lab} \nPost-molt skewness: {skew_Post_lab} ')
print(f'Pre-molt skewness: {skew_Pre_field} \nPost-molt skewness: {skew_Post_field} ')

print(f'----------------------------------------------------------------------------------------------------------------------------------------- ')

kurtosis_Pre =df[df['Location'] == 'Lab']['Pre-molt'].kurtosis()
kurtosis_Post =df[df['Location'] == 'Lab']['Post-molt'].kurtosis()
print(f'Pre-molt kurtosis for Lab data: {kurtosis_Pre} \nPost-molt kurtosis for Lab data: {kurtosis_Post} ')

kurtosis_Pre =df[df['Location'] == 'Field']['Pre-molt'].kurtosis()
kurtosis_Post =df[df['Location'] == 'Field']['Post-molt'].kurtosis()
print(f'Pre-molt kurtosis for Field data: {kurtosis_Pre} \nPost-molt kurtosis for Field data: {kurtosis_Post} ')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.read_csv('input.csv')
# Q-Q plot for normality
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
stats.probplot(df[df['Location'] == 'Lab']['Pre-molt'], dist="norm", plot=axes[0])
axes[0].set_title('Pre-Molt Size Q-Q Plot for Lab data')

stats.probplot(df[df['Location'] == 'Lab']['Post-molt'], dist="norm", plot=axes[1])
axes[1].set_title('Post-Molt Size Q-Q Plot for Lab data')

plt.show()

fig, axes = plt.subplots(1, 2, figsize=(12, 6))
stats.probplot(df[df['Location'] == 'Lab']['Pre-molt'], dist="norm", plot=axes[0])
axes[0].set_title('Pre-Molt Size Q-Q Plot for Field data')

stats.probplot(df[df['Location'] == 'Lab']['Post-molt'], dist="norm", plot=axes[1])
axes[1].set_title('Post-Molt Size Q-Q Plot for Field data')

plt.show()
```

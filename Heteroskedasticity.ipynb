```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan, het_white

# Load the data from the stored file (mp2 - input.csv)
df = pd.read_csv('input.csv')

# Linear regression model (predicting pre-molt size from post-molt size)
X = df[df['Location'] == 'Field'][['Post-molt']]
y = df[df['Location'] == 'Field']['Pre-molt']

# Add a constant term for the intercept
X_with_constant = sm.add_constant(X)

# Fit the OLS (Ordinary Least Squares) model
model = sm.OLS(y, X_with_constant).fit()

# Predictions and residuals
y_pred = model.predict(X_with_constant)
residuals = y - y_pred

# 1. Visual estimation of heteroskedasticity from residuals plot
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, color='blue', alpha=0.6)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals vs Fitted Values for Field data', fontsize=14)
plt.xlabel('Fitted Values (Predicted Pre-Molt Size)', fontsize=12)
plt.ylabel('Residuals', fontsize=12)
plt.grid(True)
plt.show()

# 2. Breusch-Pagan test
# The test requires the residuals and the independent variables (X)
bp_test = het_breuschpagan(residuals, X_with_constant)

# Extract the test statistics
bp_test_statistic, bp_p_value, bp_f_value, bp_f_p_value = bp_test

print(f'Breusch-Pagan Test Statistic: {bp_test_statistic}')
print(f'Breusch-Pagan p-value: {bp_p_value}')

# 3. White test
# The White test uses the same approach as Breusch-Pagan but allows for non-linearities
white_test = het_white(residuals, X_with_constant)

# Extract the test statistics
white_test_statistic, white_p_value, white_f_value, white_f_p_value = white_test

print(f'White Test Statistic: {white_test_statistic}')
print(f'White Test p-value: {white_p_value}')

# Linear regression model (predicting pre-molt size from post-molt size)
X = df[df['Location'] == 'Lab'][['Post-molt']]
y = df[df['Location'] == 'Lab']['Pre-molt']

# Add a constant term for the intercept
X_with_constant = sm.add_constant(X)

# Fit the OLS (Ordinary Least Squares) model
model = sm.OLS(y, X_with_constant).fit()

# Predictions and residuals
y_pred = model.predict(X_with_constant)
residuals = y - y_pred

# 1. Visual estimation of heteroskedasticity from residuals plot
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, color='blue', alpha=0.6)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals vs Fitted Values for Lab data', fontsize=14)
plt.xlabel('Fitted Values (Predicted Pre-Molt Size)', fontsize=12)
plt.ylabel('Residuals', fontsize=12)
plt.grid(True)
plt.show()

# 2. Breusch-Pagan test
# The test requires the residuals and the independent variables (X)
bp_test = het_breuschpagan(residuals, X_with_constant)

# Extract the test statistics
bp_test_statistic, bp_p_value, bp_f_value, bp_f_p_value = bp_test

print(f'Breusch-Pagan Test Statistic: {bp_test_statistic}')
print(f'Breusch-Pagan p-value: {bp_p_value}')

# 3. White test
# The White test uses the same approach as Breusch-Pagan but allows for non-linearities
white_test = het_white(residuals, X_with_constant)

# Extract the test statistics
white_test_statistic, white_p_value, white_f_value, white_f_p_value = white_test

print(f'White Test Statistic: {white_test_statistic}')
print(f'White Test p-value: {white_p_value}')
```
